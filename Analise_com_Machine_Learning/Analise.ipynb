MONTANDO MACHINE LEARNING

```python
y = df['Vendas']
x = df[['TV','Radio', 'Jornal']]
from sklearn.model_selection import train_test_split
```

```python
x_treino, x_teste,  y_treino, y_teste = train_test_split(x,y)

#x_treino, y_treino, x_teste, y_teste = train_test_split(x,y, test_size=0.3)
#divide a base de dados em 70% treino e 30% teste
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
#criar modelo regressao linear
modelo_regressaolinear = LinearRegression()

#criar modelo arvore de decisao
modelo_arvoredecisao = RandomForestRegressor()
```

```python
#treinar modelo regressao linear
modelo_regressaolinear.fit(x_treino, y_treino)
modelo_arvoredecisao.fit(x_treino, y_treino)
```

```python

#testando modelo regressao linear
previsao_regressaolinear = modelo_regressaolinear.predict(x_teste)
#testando modelo arvore de decisao
previsao_modelodecisao = modelo_arvoredecisao.predict(x_teste)
```

```python

from sklearn import metrics

print(metrics.r2_score(y_teste, previsao_regressaolinear))
print(metrics.r2_score(y_teste, previsao_modelodecisao))
```

```python
tabela = pd.DataFrame()

tabela["y_teste"] = y_teste

tabela["Previsao RegressÃ£o Linear"] = previsao_regressaolinear
tabela["Previsao Modelo DecisÃ£o"] = previsao_modelodecisao


plt.figure(figsize=(15,6))
sns.lineplot(data=tabela)
plt.show()
```
